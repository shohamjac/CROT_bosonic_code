```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib as mpl
from matplotlib import cm
from qutip import *
from qutip.ipynbtools import plot_animation
from matplotlib import animation
from IPython.display import HTML
from wigner_animation import wigner_animate
from CROT_simulation import *
```

```python
N=10
alpha = 2.0
```

```python
new = CROT_simulation('+','1',N=N)
```

```python
rho_a_final = ptrace(new.final_state, 1)
rho_a_start = ptrace(new.states[0], 1)
rho_b_final = ptrace(new.final_state, 2)
rho_b_start = ptrace(new.states[0], 2)
```

```python
plot_cavity_process(new)
```

```python
for i in range(10):
    print(fidelity(rho_a_final, ket2dm(coherent(N,-alpha*np.exp(0.1j*i))).unit()))
```

```python
new2 = CROT_simulation('+','0',N=N)
```

```python
rho_a_final2 = ptrace(new2.final_state, 1)
rho_a_start2 = ptrace(new2.states[0], 1)
rho_b_final2 = ptrace(new2.final_state, 2)
rho_b_start2 = ptrace(new2.states[0], 2)
```

```python
for i in range(10):
    print(fidelity(rho_a_final, ket2dm(coherent(N,-alpha*np.exp(0.05j*i))).unit()),
         fidelity(rho_a_final2, ket2dm(coherent(N,alpha*np.exp(0.05j*i))).unit()))
```

```python
for i in range(20):
    print(i*0.05)
    print(fidelity(rho_b_final, ket2dm(coherent(N,alpha*np.exp(0.05j*i))-coherent(N,-alpha*np.exp(0.05j*i))).unit()),
         fidelity(rho_b_final2, ket2dm(coherent(N,alpha*np.exp(0.05j*i))+coherent(N,-alpha*np.exp(0.05j*i))).unit()))
```

```python
plot_cavity_process(new2)
```
